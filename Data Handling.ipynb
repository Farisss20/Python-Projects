```python
import requests

def state_population(state):
    url = "https://jian.sh/malaysia-api/state/v1/all.json"
    response = requests.get(url)
    malaysia_state_data = response.json()

    for item in malaysia_state_data :
        if item['state'] == state:
            return item['population']
        
    return "is not found in data"
    
state=input("Enter the name of a state: ")
population = state_population(state)

if population == "state not found":
    print(population)
else:
    print("The population of " + state + " is " + str(population))
```

```python
import csv

def filter_students(input_file, output_file):
    with open(input_file, 'r', newline='') as csvfile:
        reader = csv.DictReader(csvfile)

        header = reader.fieldnames + ['Status']

        with open(output_file, 'w', newline='') as output_csvfile:
            writer = csv.DictWriter(output_csvfile, fieldnames=header)

            writer.writeheader()

            for row in reader:
                if row['Grade'] <= 'B':
                    row['Status'] = 'Passed'

                    writer.writerow(row)

input_file = 'students.csv'
output_file = 'top_students.csv'

filter_students(input_file, output_file)

print(f"Filtered data has been saved to {output_file}.")
```

```python
import csv

def read_csv(filename):
    data = []
    with open(filename, 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            data.append(row)
    return data


def filter_students(data):
    filtered_data = []
    for student in data:
        if student['Grade'] <= 'B':
            filtered_data.append(student)
    return filtered_data


def add_status_column(data):
    for student in data:
        student['Status'] = 'Passed'
    return data


def write_csv(data, output_filename):
    fieldnames = data[0].keys()
    with open(output_filename, 'w', newline='') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data)

if __name__ == "__main__":
    input_filename = "students.csv"
    output_filename = "top_students.csv"

    student_data = read_csv(input_filename)
    filtered_students = filter_students(student_data)
    students_with_status = add_status_column(filtered_students)
    write_csv(students_with_status, output_filename)

    print(f"Filtered data saved to {output_filename}")
```

```python
import csv

def filter_students(input_file, output_file):
    # Open the input CSV file
    with open(input_file, 'r', newline='') as file:
        reader = csv.DictReader(file)
        # Define headers for the output CSV file
        headers = reader.fieldnames + ['Status']
        
        # Filter students based on grade and store them in a list
        filtered_students = [row for row in reader if row['Grade'] >= 'B']
        
        # Add 'Status' column and set value to 'Passed' for filtered students
        for student in filtered_students:
            student['Status'] = 'Passed'
        
        # Write the filtered data to a new CSV file
        with open(output_file, 'w', newline='') as output:
            writer = csv.DictWriter(output, fieldnames=headers)
            writer.writeheader()
            writer.writerows(filtered_students)

# Replace 'students.csv' with your actual file name
input_file = 'students.csv'
# Replace 'top_students.csv' with the desired name for the output file
output_file = 'top_students.csv'

filter_students(input_file, output_file)
```
